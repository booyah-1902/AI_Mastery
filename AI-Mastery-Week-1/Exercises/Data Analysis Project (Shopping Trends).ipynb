```python
trends.groupby("Item Purchased").apply(lambda df: df.Location.iloc[0], include_groups=False)
# Here's one way of selecting the name of the first item purchased from each location in the dataset 
```

```python
""" 
    The code centers your data around zero by subtracting the mean from each value. This is also called "mean normalization" or "centering.
    Alternatively: trends["Review Rating"] - trends["Review Rating"].mean()
"""

trends["Review Rating"].map(lambda p: p - trends["Review Rating"].mean())
```

```python
"""
    For each combination of 'Previous Purchases' and 'Frequency of Purchases', select the customer with the highest total spending and sort all groups by 'Total Spent' in descending order.
"""

highest_spending = trends.groupby(["Previous Purchases", "Frequency of Purchases"]).apply(lambda df: df.loc[df["Total Spent"].idxmax()],include_groups=False).sort_values("Total Spent", ascending=False)
highest_spending
```

```python
type(highest_spending.index)
```

```python
highest_spenders = highest_spending.reset_index()
highest_spenders.sort_values(by='Total Spent', ascending=False) # highest_spenders.sort_index()
```

```python
highest_spenders.sort_values(by=['Gender', "Review Rating"], ascending=False)
```
